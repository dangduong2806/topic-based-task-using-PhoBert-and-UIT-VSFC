```python
import pandas as pd
import json
import os

def augment_data(original_file, generated_file, output_file):
    # 1. Load d·ªØ li·ªáu g·ªëc
    try:
        df_train = pd.read_csv(original_file)
        print(f"‚úÖ D·ªØ li·ªáu g·ªëc: {len(df_train)} d√≤ng.")
    except FileNotFoundError:
        print(f"‚ùå Kh√¥ng t√¨m th·∫•y file g·ªëc {original_file}")
        return

    # 2. Load d·ªØ li·ªáu sinh th√™m t·ª´ ChatGPT
    try:
        with open(generated_file, "r", encoding="utf-8") as f:
            new_data = json.load(f)
        
        df_new = pd.DataFrame(new_data)
        # --- FIX L·ªñI: T·ª∞ ƒê·ªòNG ƒê·ªîI T√äN C·ªòT V·ªÄ 'sents' ---
        print(f"üîç C√°c c·ªôt t√¨m th·∫•y trong JSON: {list(df_new.columns)}")
        
        # --- B·ªò L·ªåC CH·∫§T L∆Ø·ª¢NG (FILTRATION) ---
        
        # L·ªçc 1: Lo·∫°i b·ªè c√¢u qu√° ng·∫Øn (r√°c)
        df_new = df_new[df_new['text'].str.len() > 5] 
        
        # L·ªçc 2: Lo·∫°i b·ªè c√¢u tr√πng l·∫∑p y h·ªát trong t·∫≠p train c≈©
        existing_sents = set(df_train['text'].str.lower().str.strip())
        # Ch·ªâ gi·ªØ l·∫°i c√¢u m√Ä t·∫≠p train c≈© CH∆ØA c√≥
        df_new = df_new[~df_new['text'].str.lower().str.strip().isin(existing_sents)]
        
        print(f"‚úÖ D·ªØ li·ªáu sinh th√™m h·ª£p l·ªá: {len(df_new)} d√≤ng.")
        
        # Hi·ªÉn th·ªã v√†i m·∫´u ƒë·ªÉ check ƒë·ªô 'ch·∫•t'
        print("\n--- M·∫´u d·ªØ li·ªáu 'ƒë·∫≠m ch·∫•t UIT' v·ª´a sinh ---")
        print(df_new[['text', 'label']].head(5))
        print("--------------------------------------------\n")

    except Exception as e:
        print(f"‚ùå L·ªói khi ƒë·ªçc file JSON: {e}")
        return

    # 3. Gh√©p v√† Shuffle
    # ƒê·∫£m b·∫£o c·ªôt kh·ªõp nhau
    required_cols = ['text', 'label']
    df_final = pd.concat([df_train[required_cols], df_new[required_cols]], ignore_index=True)
    
    # X√°o tr·ªôn d·ªØ li·ªáu (Shuffle) ƒë·ªÉ model kh√¥ng h·ªçc theo th·ª© t·ª±
    df_final = df_final.sample(frac=1, random_state=42).reset_index(drop=True)
    
    # 4. L∆∞u file
    df_final.to_csv(output_file, index=False)
    print(f"üéâ ƒê√£ t·∫°o file dataset m·ªõi: {output_file} ({len(df_final)} d√≤ng)")
    print("üëâ H√£y c·∫≠p nh·∫≠t config.yaml ƒë·ªÉ tr·ªè v√†o file n√†y!")

# --- CH·∫†Y ---
if __name__ == "__main__":
    augment_data(
        original_file="D:/UIT-VSFC/data/processed/train.csv",       # File g·ªëc c·ªßa b·∫°n
        generated_file="uit_generated_others.json", # File copy t·ª´ ChatGPT
        output_file="D:/UIT-VSFC/data/processed/train_augmented.csv" # File k·∫øt qu·∫£
    )
```
