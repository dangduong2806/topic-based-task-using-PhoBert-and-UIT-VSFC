```python
import pandas as pd
import json
import os

def augment_data(original_file, generated_file, output_file):
    # 1. Load dữ liệu gốc
    try:
        df_train = pd.read_csv(original_file)
        print(f"✅ Dữ liệu gốc: {len(df_train)} dòng.")
    except FileNotFoundError:
        print(f"❌ Không tìm thấy file gốc {original_file}")
        return

    # 2. Load dữ liệu sinh thêm từ ChatGPT
    try:
        with open(generated_file, "r", encoding="utf-8") as f:
            new_data = json.load(f)
        
        df_new = pd.DataFrame(new_data)
        # --- FIX LỖI: TỰ ĐỘNG ĐỔI TÊN CỘT VỀ 'sents' ---
        print(f"🔍 Các cột tìm thấy trong JSON: {list(df_new.columns)}")
        
        # --- BỘ LỌC CHẤT LƯỢNG (FILTRATION) ---
        
        # Lọc 1: Loại bỏ câu quá ngắn (rác)
        df_new = df_new[df_new['text'].str.len() > 5] 
        
        # Lọc 2: Loại bỏ câu trùng lặp y hệt trong tập train cũ
        existing_sents = set(df_train['text'].str.lower().str.strip())
        # Chỉ giữ lại câu mÀ tập train cũ CHƯA có
        df_new = df_new[~df_new['text'].str.lower().str.strip().isin(existing_sents)]
        
        print(f"✅ Dữ liệu sinh thêm hợp lệ: {len(df_new)} dòng.")
        
        # Hiển thị vài mẫu để check độ 'chất'
        print("\n--- Mẫu dữ liệu 'đậm chất UIT' vừa sinh ---")
        print(df_new[['text', 'label']].head(5))
        print("--------------------------------------------\n")

    except Exception as e:
        print(f"❌ Lỗi khi đọc file JSON: {e}")
        return

    # 3. Ghép và Shuffle
    # Đảm bảo cột khớp nhau
    required_cols = ['text', 'label']
    df_final = pd.concat([df_train[required_cols], df_new[required_cols]], ignore_index=True)
    
    # Xáo trộn dữ liệu (Shuffle) để model không học theo thứ tự
    df_final = df_final.sample(frac=1, random_state=42).reset_index(drop=True)
    
    # 4. Lưu file
    df_final.to_csv(output_file, index=False)
    print(f"🎉 Đã tạo file dataset mới: {output_file} ({len(df_final)} dòng)")
    print("👉 Hãy cập nhật config.yaml để trỏ vào file này!")

# --- CHẠY ---
if __name__ == "__main__":
    augment_data(
        original_file="D:/UIT-VSFC/data/processed/train_augmented.csv",       # File gốc của bạn
        generated_file="uit_generated_others_2.json", # File copy từ ChatGPT
        output_file="D:/UIT-VSFC/data/processed/train_augmented_2.csv" # File kết quả
    )
```
